# Seaborn Penguins Dataset Exploration

This notebook covers 10 questions using the Seaborn `penguins` dataset with different plot types.

```python
import seaborn as sns
import pandas as pd
import numpy as np

df = sns.load_dataset('penguins')
```

```python
df.head()
```

## 1. Average flipper length for each species (Barplot)

```python
print(df.groupby('species')['flipper_length_mm'].mean())
```

```python
sns.barplot(data=df, x='species', y='flipper_length_mm', estimator='mean', errorbar='ci')
```

## 2. Body mass by island (Boxplot)

```python
sns.boxplot(data=df, x='island', y='body_mass_g')
```

```python
sns.barplot(data=df, x='species', y='body_mass_g', estimator='sum')
```

## 3. Count of each species (Countplot)

```python
sns.countplot(data=df,x='species')
```

## 4. Flipper length by species and gender (barplot with Hue)

```python
sns.barplot(data=df, x='species', y='flipper_length_mm', estimator='mean', hue='sex')
```

```python
## 4. Flipper length by species and gender (barplot with Hue)
sns.barplot(data=df, x='species', y='bill_length_mm', estimator='mean', hue='sex')
```

## 5. Pairplot of numerical features colored by species

```python
sns.pairplot(data=df, hue='species')
```

## 6. Heatmap of correlations between numerical features

```python
corr_df=df.corr(numeric_only=True)
```

```python
sns.heatmap(corr_df, cmap='coolwarm')
```

## 7. Average body mass by species and gender (Barplot with Hue)

```python
sns.barplot(data=df, x='species', y='body_mass_g', estimator='mean', hue='sex')
```

## 8. Stripplot of flipper length by species

```python
sns.stripplot(data=df, x='species', y='flipper_length_mm')
```

```python
sns.swarmplot(data=df, x='species', y='flipper_length_mm')
```

## 10. Flipper length vs body mass by species (FacetGrid + scatterplot)

```python
g = sns.FacetGrid(df,row='sex' ,col='species')
g.map_dataframe(sns.scatterplot, x='flipper_length_mm', y='body_mass_g')
```
